```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import plotly.express as px
```

```python
# Grouping by week and summing sales per week
grouped_data = weekly_ticket_sales_df.groupby('Week')['TicketsSold'].sum()

# Creating a line plot
plt.figure(figsize=(10, 6))
plt.plot(grouped_data.index, grouped_data.values, marker='o')
plt.xlabel('Week')
plt.ylabel('Total Sales per Week')
plt.title('Total Ticket Sales per Week')
plt.grid(True)
plt.show()
```

```python
# plot the the graph of ticket sold per week for event with the most number of weeks
# Group by 'eventName' and count unique weeks
event_week_counts = weekly_ticket_sales_df.groupby('EventName')['Week'].nunique()

# Find the event with the most number of weeks
event_with_most_weeks = event_week_counts.idxmax()

# Filter DataFrame for the event with the most weeks
selected_event_df = weekly_ticket_sales_df[weekly_ticket_sales_df['EventName'] == event_with_most_weeks]

# Plot 'ticketsold' per week for the selected event
plt.figure(figsize=(10, 6))
plt.plot(selected_event_df['Week'], selected_event_df['TicketsSold'], marker='o')
plt.xlabel('Week')
plt.ylabel('Tickets Sold')
plt.title(f'Tickets Sold per Week for Event: {event_with_most_weeks}')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# plot the the graph of ticket sold per week for event with the most tickets sold
# Step 1: Group by 'eventName' and sum the 'ticket sold' values
event_total_tickets = weekly_ticket_sales_df.groupby('EventName')['TicketsSold'].sum()

# Step 2: Find the event with the most total tickets sold
event_with_most_tickets = event_total_tickets.idxmax()

# Step 3: Filter DataFrame for the event with the most tickets sold
selected_event_df = weekly_ticket_sales_df[weekly_ticket_sales_df['EventName'] == event_with_most_tickets]

# Step 4: Plot 'ticket sold' per week for the selected event
plt.figure(figsize=(8, 4))
plt.plot(selected_event_df['Week'], selected_event_df['TicketsSold'], marker='o')
plt.xlabel('Week')
plt.ylabel('Ticket Sold')
plt.title(f'Tickets Sold per Week for Event: {event_with_most_tickets}')
plt.xticks(rotation=45)
plt.tight_layout()

plt.show()
```

```python
pattern_df= weekly_ticket_sales_df.copy()
pattern_df
```

```python
#add the event type column from Org1_df to pattern df
# Create an empty list to store the EventType values
event_types = []

# Iterate through each row in pattern_df
for index, row in pattern_df.iterrows():
    event_id = row['EventId']
    matching_row = org1_df[org1_df['EventId'] == event_id]
    
    if not matching_row.empty:
        event_type = matching_row['EventType'].values[0]
        event_types.append(event_type)
    else:
        event_types.append(None)  # or some default value if no match is found

# Add the event_types list as a new column in pattern_df
pattern_df['EventType'] = event_types
pattern_df
```

```python
pattern_df.info()
```

```python
# plot of ticket sales comparison by event type
desired_event_types = ["Conference / Symposium", "Seminar", "Research Training", "Lecture"]

# Filter the DataFrame for the desired event types
filtered_df = pattern_df[pattern_df['EventType'].isin(desired_event_types)]

# Create a pivot table to organize data for plotting
pivot_table = filtered_df.pivot_table(index='Week', columns='EventType', values='TicketsSold', aggfunc='sum')

# Create a figure and axes with specified size
fig, ax = plt.subplots(figsize=(10, 6))  # Adjust the width and height as needed

# Plotting
pivot_table.plot(kind='area', stacked=True, ax=ax)

ax.set_xlabel('Weeks')
ax.set_ylabel('Number of Tickets Sold')
ax.set_title('Ticket Sales Comparison by Event Type')
ax.legend(title='Event Type')
plt.tight_layout()
plt.show()
```

## Org 1 dataset visualization

```python
# visualiztion for org 1
org1_df=pd.read_csv('ORG01-upd.csv')
org1_df.head()
```

```python
org1_df['EventType'].value_counts()
```

```python
# Create a distplot for EventType column
sns.set(style="whitegrid")
ax=sns.countplot(data=org1_df, x=org1_df['EventType'])

# Customize x-axis tick labels
ax.set_xticklabels(ax.get_xticklabels(), rotation=45, ha="right")

# Adding labels and title
plt.xlabel('EventType')
plt.ylabel('Count')
plt.title('Distribution of Event Type')

# Display the plot
plt.show()
```

From the distribution plot and the output of the code above, the event type Seminar is the most occuring with 36,377 counts followed by Reserch Training with 6,844 counts.

```python
# Create a pie chart for EventType column
event_type_counts = org1_df['EventType'].value_counts()
labels = event_type_counts.index
sizes = event_type_counts.values

# Create the pie chart
plt.figure(figsize=(10, 6))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140, pctdistance=0.85,
        colors=sns.color_palette('pastel'), wedgeprops={'edgecolor': 'black'})

# Draw circle to make pie chart circular
centre_circle = plt.Circle((0,0),0.70)
fig = plt.gcf()


# Equal aspect ratio ensures that pie is drawn as a circle
plt.tight_layout()
plt.title('Distribution of Event Type')
plt.show()
```
